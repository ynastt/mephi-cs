### Решение

### Вычисляем частные производные функции $ u(x, y, z) $:

- По $ z $:  
$
\frac{\partial u}{\partial z} = 2z - 2
$

### Условие стационарных точек
Найдём точки, где все частные производные равны нулю:

$
\frac{\partial u}{\partial x} = 0, \quad \frac{\partial u}{\partial y} = 0, \quad \frac{\partial u}{\partial z} = 0
$

Для этого решаем систему уравнений:

$
9x^2 + 6y = 0 \quad \text{(1)}
$

$
2y + 6x = 0 \quad \text{(2)}
$

$
2z - 2 = 0 \quad \text{(3)}
$

Из (3) получаем $z=1$.  
Остатеся система из (1) и (2).

Из (1): $ y = -\frac{9x^2}{6} \ ⇒  y = -\frac{3}{2}x^2 $

Из (2) $ y = -\frac{6x}{2} \ ⇒ y = -3x$


Сравним $ y = -\frac{3}{2}x^2 $ и $ y = -3x $ и найдем $x$ точек:

$
-\frac{3}{2}x^2 = -3x \quad \Rightarrow  \quad \frac{3}{2}x^2 = 3x \quad \Rightarrow\quad \frac{3}{2}x^2 - 3x = 0 \quad \Rightarrow \quad \frac{3}{2}x(x - 2) = 0
$

Отсюда получаем:
$
x = 0 \quad \text{или} \quad x = 2
$


### Находим критические (стационарные) точки
- При $ x = 0 $:  
$
y = -3 \cdot 0 = 0, \quad z = 1 \quad \Rightarrow \quad (x, y, z) = (0, 0, 1)
$  

- При $ x = 2 $:  
$
y = -3 \cdot 2 = -6, \quad z = 1 \quad \Rightarrow \quad (x, y, z) = (2, -6, 1)
$

### Исследуем критические точки на экстремумы
Для исследования характера стационарных точек используем матрицу вторых частных производных (матрицу Гессе)

Находим вторые производные:  

$
\frac{\partial^2 u}{\partial x^2} = 18x

$
\frac{\partial^2 u}{\partial y^2} = 2

Матрица Гессе:
$
H =

#### Проверка для точки $ (0, 0, 1) $:
$
H =
\begin{pmatrix}

Используем критерий Сильвестра, считаем старшие миноры матрицы.  

Старший минор 1-го порядка:
$
\Delta_1 = 0
$

Старший минор 2-го порядка:

\Delta_2 =

$

Старший минор 3-го порядка (определитель матрицы):
$
\Delta_3 =

$

Так как миноры не знакочередующиеся и не положительны, точка $(0, 0, 1)$ — седловая. Не является экстремумом функции.

---

#### Проверка для точки $ (2, -6, 1) $:

H =

Старший минор 3-го порядка (определитель матрицы):

Так как все главные миноры положительны, точка $(2, -6, 1)$ — точка минимума

# Задание 3  

Найдите условный экстремум функции:  
$z=e^{-x-y}$  
Если задано следующее условие:  
$\frac{x^2}{4}+y^2=5$

Подробно распишите решение. Задание должно быть решено без использования Python.

### Решение

Для решения задачи используем метод множителей Лагранжа.

## Запишем функцию Лагранжа

$
\mathcal{L}(x, y, \lambda) = e^{-x-y} + \lambda \left( \frac{x^2}{4} + y^2 - 5 \right)
$

## Вычисляем частные производные функции Лагранжа:
$
\frac{\partial \mathcal{L}}{\partial x} = -e^{-x-y} + \lambda \cdot \frac{x}{2}
$

$
\frac{\partial \mathcal{L}}{\partial y} = -e^{-x-y} + 2\lambda y
$

$
\frac{\partial \mathcal{L}}{\partial \lambda} = \frac{x^2}{4} + y^2 - 5
$

Далее приравниваем производные к нулю и решаем полученную систему, чтобы найти критические точки.

$
\frac{\partial \mathcal{L}}{\partial x} = 0 \quad \text{(1)}
$

$
\frac{\partial \mathcal{L}}{\partial y} = 0 \quad \text{(2)}
$

$
\frac{\partial \mathcal{L}}{\partial \lambda} = 0 \quad \text{(3)}
$

## Решаем систему уравнений:

Из уравнения (1):

$
\lambda = \frac{2 e^{-x-y}}{x} \quad (x \neq 0)
$

Из (2):

$
\lambda = \frac{e^{-x-y}}{2y} \quad (y \neq 0)
$

Приравниваем полученные выражения для $\lambda$:

$
\frac{2 e^{-x-y}}{x} = \frac{e^{-x-y}}{2y}
$

Упрощаем уравнение:

$ e^{-x - y} \neq 0 ⇒ $ можно сократить. Получаем:  

$
\frac{2}{x} = \frac{1}{2y}
$

Отсюда получаем $ 4y = x $

подставляем $x = 4y$ в условие ограничения $\left( \frac{x^2}{4} + y^2 = 5 \right)$:


$
\frac{(4y)^2}{4} + y^2 = 5
$

$
\frac{16y^2}{4} + y^2 = 5
$

Упрощаем:  

$
4y^2 + y^2 = 5
$

$
5y^2 = 5 \quad \Rightarrow \quad y^2 = 1 \quad \Rightarrow \quad y = \pm 1
$

Находим соответствующие значения $x$:  

- Для $y = 1$:  
 $x = 4 \cdot 1 = 4$.

- Для $y = -1$:  
$x = 4 \cdot (-1) = -4$.

Подставим точки в $z = e^{-x-y}$ и найдем соответствующие значения функции $z(x,y)$:

- Для $(x, y) = (4, 1)$:  
$
z = e^{-4-1} = e^{-5}
$

- Для $(x, y) = (-4, -1)$:  
$
z = e^{-(-4) - (-1)} = e^{4 + 1}= e^{5}
$

## Ответ
Условный экстремум функции $z = e^{-x-y}$ при условии $\frac{x^2}{4} + y^2 = 5$ достигается в следующих точках:
- минимум в точке (4,1) со значением $e^{-5}$,   
- максимум в точке (-4, -1) со значением $e^{5}$.

# Задание 4

Вам предоставлены данные о людях с наличием или отсутствием заболевания: data_exam.csv

Признаки:

- Age — возрастная группа.
- Test — позитивный или негативный тест на заболевание.
- Status — целевая переменная, есть инфекция или нет.

Необходимо реализовать алгоритм Наивного Байеса для решения задачи классификации. Обязательно оцените качество полученного результата по итогу.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

data = pd.read_csv("data_exam.csv")
data.head()
```

```python
# Преобразуем категориальные признаки в числовые
encoder = LabelEncoder()

data['Test'] = encoder.fit_transform(data['Test'])
data['Age_Group'] = encoder.fit_transform(data['Age_Group'])
data['Status'] = encoder.fit_transform(data['Status'])

# Создаем признаки и целевую переменную
X = data[['Age_Group', 'Test']]
y = data['Status']

# Разделяем на обучающую и тестовую выборки данные (75% на обучение, 25% на тест)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Создаем и обучаем модель Наивного Байесовского классификатора
model = GaussianNB()
model.fit(X_train, y_train)

# Получаем предсказание на тестовой выборке
y_pred = model.predict(X_test)

# Оцениваем качество модели
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}\n')
conf_matrix = confusion_matrix(y_test, y_pred)
print(f'Матрица ошибок:\n{conf_matrix}\n')
# классификационный отчет
# Здесь важно посмотреть метрики: precision, recall, f1-score
class_report = classification_report(y_test, y_pred)
print('Classification Report:')
print(class_report)
```

**Вывод:**  
Модель показала неплохую производительность. Значение метрики F1-Score (Среднее гармоническое PRECISION и RECALL) показывает, что модель хорошо справляется с классификацией людей с отсутсвием инфекции (0.71 > 0.5) и не очень точно с классификацией людей с инфекцией (0.55, что не так далеко от 0.5). Точность Accuracy = 0.64, следовательно предсказания модели больше правильные, чем неправильные. Тем не менее, производительность и качество модели были бы лучше, если бы выборка была больше при условии таких же сбалансированных классов.

# Задание 5

Дана функция:  
$ f(x) = x^4 + 3x^3 - 12x^2 + 7x -2 $  


Оптимизируйте её с помощью метода Ньютона, используя язык программирования Python.

За начальную точку возьмите х = 10, необходимую точность 0.0001.

```python
import sympy as sp

# переменнуая х и функция f(x) через символы
x = sp.symbols("x")
f = x**4 + 3 * x**3 - 12 * x**2 + 7 * x - 2
print("Функция: ", f)

# первая и вторая производные
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)
print("Первая производная: ", f_prime)
print("Вторая производная: ", f_double_prime)

# конвертируем sympy выражения производных в функции для вычислений
f_prime_func = sp.lambdify(x, f_prime, "numpy")
f_double_prime_func = sp.lambdify(x, f_double_prime, "numpy")

# метод Ньютона
def newton_method(x0, eps=0.0001, max_iter=1000):
    x = x0

    for i in range(max_iter):

        f_prime_val = f_prime_func(x)
        f_double_prime_val = f_double_prime_func(x)

        if f_double_prime_val == 0:
            print("Ошибка: вторая производная равна нулю.")
            return None

        # считаем новую точку х
        x_new = x - f_prime_val / f_double_prime_val

        # проверяем сходимость
        if abs(x_new - x) < eps:
            # Найдено оптимальное значение
            return x_new

        # обновляем точку х
        x = x_new

    print("Метод не сошелся за указанное максимальное количество итераций")
    return None


x0 = 10
res = newton_method(x0, eps=0.0001)
print("\nРезультаты метода Ньютона")
print(f"Оптимальное значение x: {res}")
print(f"Значение функции в оптимальной точке: {f.subs(x, res)}")
```

Дано распределение случайной величины, которая отражает вероятность получения разного количества спам-писем в течение дня:

2. Найдите математическое ожидание для количества полученных писем.  
3. Найдите дисперсию для количества полученных писем. Ответ округлите до сотых.

Задание должно быть решено без использования готовых функций Python.

## Решение

### 1. Вероятность получения 7 спам-писем.

Сумма всех вероятностей для дискретного распределения равна 1, следовательномы можем вычислить вероятность для $X = 7$:
$P(X=7) = 1 - P(X=0) - P(X=1) - P(X=2) - P(X=3) - P(X=4) - P(X=5) - P(X=6)$
$P(X=7 ) = 1 - 0.1 - 0.3 - 0.1 - 0.25 - 0.05 - 0.05 - 0.1$  

**Ответ:**
$P(X=7) = 0.05$

### 2. Математическое ожидание для количества полученных писем.  

Мат. ожидание для дискретного распределения рассчитывается по формуле:
$E(X) = \sum_{i} X_i \cdot P(X_i)$
где $X_i$ — значение случайной величины, а $P(X_i)$ — соответствующая ей вероятность. В нашем случае $i \in [0, 7]$

Считаем:
$E(X) = 0 \cdot P(X=0) + 1 \cdot P(X=1) + 2 \cdot P(X=2) + 3 \cdot P(X=3) + 4 \cdot P(X=4) + 5 \cdot P(X=5)= 0 \cdot 0.1 + 1 \cdot 0.3 + 2 \cdot 0.1 + 3 \cdot 0.25 + 4 \cdot 0.05 + 5 \cdot 0.05 + 6 \cdot 0.1 + 7 \cdot 0.05 = 0 + 0.3 + 0.2 + 0.75 + 0.2 + 0.25 + 0.6 + 0.35 = 2.65$

**Ответ:**
$E(X) = 2.65$

### 3. Дисперсия для количества полученных писем. Ответ округлите до сотых.

Дисперсия  для дискретного распределения рассчитывается по формуле:
$D(X) = \sum_{i}X_i^2\cdot P(X_i) - E(X)^2 $
где $X_i$ — значение случайной величины, $P(X_i)$ — соответствующая ей вероятность, а $E(X)$ — мат ожидание (среднее значение случайной величины). В нашем случае $i \in [0, 7]$

$D(X) = (0^2 \cdot 0 + 1^2 \cdot 0.3 + 2^2 \cdot 0.1 + 3^2 \cdot 0.25 + 4^2 \cdot 0.05 + 5^2 \cdot 0.05 + 6^2 \cdot 0.1 + 7^2 \cdot 0.05 ) - 2.65^2 = 0 + 0.3 + 0.4 + 2.25 + 0.8 + 1.25 + 3.6 + 2.45 - 2.65^2 =11.05 - 7.0225 = 4.0275$

**Ответ:**
$D(X) = 4.0275$

## Решение  

Обозначим случауную величину $X$ - поступление звонка в колл-центр.   
Надо найти, что за час в колл-центр поступит не более 4 и не менее 2 звонков: $P(2 \leq X \leq 4) = P(X=2) + P(X=3) + P(X=4) $

В задаче дана частота случайного события $X$ - поступление звонка в колл-центр: в среднем 16 звонков за 8 часов. Это задача на распределение Пуассона.

Запишем формулу для распределения Пуассона:
$$P(X=k) = \frac{\lambda^k \ e^{-\lambda}}{k!}$$

где $k$ - число событий, $\lambda$ - среднее количество событий за фиксированный промежуток времени.

В наше задаче $k = 2, 3, 4$. А также мы знаем среднее число звонков за 8 часов, обозначим $\lambda_8=16$.

Находим среднее число звонков за 1 час:
$\lambda_1 = \frac{\lambda_8}{8}=\frac{16}{8}=2$

Теперь с учетом $\lambda_1$ вычислим $P(X=2), P(X=3), P(X=4)$ и посчитаем искомую вероятность.

$P(X=2) = \frac{2^2 \ e^{-2}}{2!}$

$P(X=3) = \frac{2^3 \ e^{-2}}{3!}$

$P(X=2) = \frac{2^4 \ e^{-2}}{4!}$

$P(2 \leq X \leq 4) \approx 0.541$

Воспользуемся кодом (без готовых функций, за исключением экспоненты и факториала) ниже для проверки

```python
import math
lambd = 2

res = 0
for k in range(2, 5):
    p = (lambd**k * math.exp(-lambd)) / math.factorial(k)
    res += p
    print(f"P(X={k}) = {p}")

print(f"P(2≤X≤4) = {res:.4f}")
```
