**Задача 1**. (11 баллов)

В известном магазине техники представлено очень много экземпляров различной бытовой техники двух видов: с возможностью подключения к системе "Умный дом" и без такой возможности. Причем, известно, что эти два вида техники представлены в пропорции 1 : 3 соответственно. Случайным образом выбрали 7 экземпляров техники для розыгрыша среди участниц конкурса "Мисс Рукоделье - 2024".  Какова вероятность, что среди них не более двух экземпляров с возможностью подключения к системе "Умный дом"?

Подробно распишите решение и обоснуйте выбор распределения для моделирования сюжета задачи. Ответ округлите до трёх знаков.


**Выбор распределения**  
Для моделирования сюжета задачи используется биномиальное распределение, поскольку у нас имеется два исхода для каждого экземпляра (с возможностью подключения к системе "Умный дом" или без),  событяе выбора независиыее а вероятность успеха (выбрали с возможностью умного дома) одинакова в каждом испытании.

Поскольку техника с возможностью подключения к системе "Умный дом"  и без этой возможности представлены в пропорции 1 к 3, то вероятность того, что выбранный экземпляр будет с возможностью подключения к системе "Умный дом" (т.н. вероятность успеха), равна:
$p = 1/4 = 0.25$
Соответственно, вероятность отсутствия возможности подкл.чения к "Умному дому" (т.н. вероятность неудачи) равна $q = 1-p = 3/4 = 0.75$.  
Пусть $X$ - это случайное событие выбора экземпляра техники с возможностью подключения к умному дому.  
$P\{X\leq2\}=P\{X=0\}+P\{X=1\}+P\{X=2\}=C_7^0p^0q^7+C_7^1p^1q^6+C_7^2p^2q^5 \approx 0.756$

```python
from scipy.stats import binom

n = 7 # количество испытаний
p = 0.25 # вероятность, что в случайном испытании будет устройство с Умным домом
k = 2 # ожидаемое количество успехов
# для вычисления используем кумулятивную функцию для биномиального распределения
print('P{среди выбранных 7 экземпляров техники не более двух с возможностью подключения к системе "Умный дом"}=', round(binom.cdf(k, n, p), 3))
```

**Задача 2** (30 баллов)

Дан закон распределения количества публикаций в которой группе телеграм-каналов:

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline X \text { (число публикаций) } & 0 & 1 & 2 & 3 & 4 & 5 \\
\hline \mathrm{P} \text { (вероятность) } & ? & 0,3 & 0,3 & 0,2 & 0,1 & 0,1 \\
\hline
\end{array}
$$

Найдите пропущенную вероятность для X = 0, математическое ожидание и дисперсию. Решите задачу "вручную" и с помощью python.

1) Пропущенная вероятность для X=0
Мумма всех вероятностей для дискретного распределения равна 1, следовательномы можем вычислить вероятность для $X = 0$:
$P(X=0) = 1 - P(X=1) - P(X=2) - P(X=3) - P(X=4) - P(X=5)$
$P(X=0) = 1 - 0.3 - 0.3 - 0.2 - 0.1 - 0.1$  
$P(X=0) = 0$

2) Математическое ожидание
Мат. ожидание для дискретного распределения рассчитывается по формуле:
$E(X) = \sum_{i} X_i \cdot P(X_i)$
где $X_i$ — значение случайной величины, а $P(X_i)$ — соответствующая ей вероятность.

Считаем:
$E(X) = 0 \cdot P(X=0) + 1 \cdot P(X=1) + 2 \cdot P(X=2) + 3 \cdot P(X=3) + 4 \cdot P(X=4) + 5 \cdot P(X=5)= 0+ 1 \cdot 0.3 + 2 \cdot 0.3 + 3 \cdot 0.2 + 4 \cdot 0.1 + 5 \cdot 0.1=0.3 + 0.6 + 0.6 + 0.4 + 0.5 = 2.4$

3) Дисперсия
Дисперсия  для дискретного распределения рассчитывается по формуле:
$D(X) = \sum_{i}X_i^2\cdot P(X_i) - E(X)^2 $

$D(X) = 0^2 \cdot 0 + 1^2 \cdot 0.3 + 2^2 \cdot 0.3 + 3^3 \cdot 0.2 + 4^2 \cdot 0.1 + 5^2 \cdot 0.1 = 0.3 + 1.2 + 1.8 + 1.6 + 2.5 - 2.4^2 = 1.64$

```python
import numpy as np

X = np.array([0, 1, 2, 3, 4, 5])
P = np.array([0, 0.3, 0.3, 0.2, 0.1, 0.1])

P0 = 1 - np.sum(P)
print('Пропущенная вероятность для X = 0:', P0)

mean = np.sum(X * P)
print('Математическое ожидание:', mean)

var = np.sum((X - mean)**2 * P)
print('Дисперсия:', round(var, 4))
```

**Задача 3** (25 баллов)


Компания приобрела 10 000 новых ноутбуков. Вероятность поломки одного ноутбука в течение месяца равна 0,0002. Предполагается, что поломки отдельных ноутбуков происходят независимо друг от друга.

**Вопрос 1**. Найдите вероятность того, что за месяц сломается ровно 1 ноутбук.
Решение объясните. (5 баллов)


Это задача на биномиальное распределение, поскольку поломки ноутбуков происходят независимо друг от друга (10000 независимых испытаний), вероятность поломки одного ноутбука в течение месяца (т.н. вероятность успеха) равна 0,0002 и не изменяется.

```python
# Вопрос 1.
n = 10000 # количество испытаний
p = 0.0002 # вероятность 1поломки ноутбука в течение месяца
k = 1 # ожидаемое количество успехов

print('P{За месяц сломается ровно 1 ноутбук}=', binom.pmf(k, n, p))
```

**Вопрос 2**. Компания планирует ввести программу расширенной гарантии, по которой она будет бесплатно ремонтировать или заменять ноутбуки, сломавшиеся в течение месяца. Если общее количество поломок за месяц превысит 5 ноутбуков, компания понесет значительные дополнительные расходы.

а) Найдите вероятность того, что за месяц сломается более 5 ноутбуков.Решение объясните (10 баллов)

б) Если стоимость ремонта или замены одного ноутбука составляет 500 долларов, вычислите математическое ожидание общих расходов компании на поломки за один месяц. (10 баллов)

```python
# Вопрос 2.а)
# X - случайное событие поломки ноутбука
# P(X>5) = 1 - P(X <= 5)
# Используем binom.sf(5, n, p), которая вычисляет обратную кумулятивную вероятность (survival function), т.е. 1 - P(X <= 5) 
from scipy.stats import binom

P5 = binom.sf(5, n, p)
print('P{за месяц сломается более 5 ноутбуков}=', P5)
```

```python
# Вопрос 2.б)
# Для того, чтобы найти математическое ожидание общих расходов компании на поломки за 1 месяц,
# нужно математическое ожидание количества поломок умножить на стоимость ремонта
# найдем матожидание количества поломок
price = 500 # стоимость ремонта или замены одного ноутбука 

# математическое ожидание количества поломок
M_count = n * p

# математическое ожидание общих расходов компании на поломки
total = M_count * price
print('Математическое ожидание общих расходов компании на поломки:', total)
```
