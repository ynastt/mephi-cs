1. Напишите запрос, который позволит узнать, какие работодатели находятся на первом и пятом месте по количеству вакансий.

```python
# текст запроса
query_5_1 = f'''
    WITH employer_vacancies AS (
        SELECT
            e.name AS employer,     -- работодатель
            COUNT(v.id) AS cnt      -- количество вакансий
        FROM employers e
        JOIN vacancies v
            ON e.id = v.employer_id
        GROUP BY e.name            -- группируем по названию работодателя
    ),
    ranked_employers AS (
        SELECT
            employer,
            cnt,
            DENSE_RANK() OVER (ORDER BY cnt DESC) AS rank  -- ранжируем по числу вакансий
        FROM employer_vacancies
    )
    SELECT
        employer,                  -- работодатель
        cnt,                       -- количество вакансий
        rank                       -- ранг (место) по количеству вакансий
    FROM ranked_employers
    WHERE rank IN (1, 5)            -- берем 1 и 5 место
    ORDER BY rank;                  -- сортируем по месту
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_5_1, connection)
df
```

```python
# код для получения списка городов-милионников
url = "https://ru.wikipedia.org/wiki/Города-миллионеры_России"

headers = {
    "User-Agent": "Mozilla/5.0 (Colab; pandas read_html)"
}

resp = requests.get(url, headers=headers, timeout=30)
resp.raise_for_status()
resp.encoding = "utf-8"

html = resp.text
tables = pd.read_html(StringIO(html))

# Берём таблицу, где есть колонка "Город"
df = next(t for t in tables if "Город" in t.columns)

million_cities = (
    df["Город"]
    .astype(str)
    .str.replace(r"\[.*?\]", "", regex=True)
    .str.replace(r"\(.*?\)", "", regex=True)
    .str.strip()
    .tolist()
)

cities = ", ".join("'" + c.replace("'", "''") + "'" for c in million_cities)
print(cities)
```

```python
# текст запроса
query_5_7 = f'''
    WITH yandex AS (
        SELECT
            a.name AS city,               -- название города
            COUNT(v.id) AS vacancies_cnt  -- количество вакансий Яндекса в городе
        FROM vacancies v
        JOIN employers e                  -- соединяем вакансии с таблицей работодателей
            ON v.employer_id = e.id
        JOIN areas a                      -- соединяем вакансии с таблицей регионов
            ON v.area_id = a.id
        WHERE e.name = 'Яндекс'           -- оставляем только для Яндекса
        AND a.name IN ({cities})      -- только города-миллионники
        GROUP BY a.name                   -- группируем по городам
    )

    SELECT
        city,  -- город
        vacancies_cnt    -- количество вакансий
    FROM yandex

    UNION ALL

    SELECT
        'Total' AS city, -- добавляем строку с общим количеством вакансий компании
        SUM(vacancies_cnt) AS vacancies_cnt
    FROM yandex

    ORDER BY vacancies_cnt ASC;   -- сортируем по возрастанию количества вакансий
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_5_7, connection)
df
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в среднем указывают для каждого типа требуемого опыта (уникальное значение из поля `experience`).

    При решении задачи примите во внимание следующее:
    1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
    2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
    3. Если в расчётах участвует `null`, в результате он тоже даст `null` (посмотрите, что возвращает запрос `select 1 + null`). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит `null` на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

    Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до **целого числа**.

```python
# текст запроса
query_6_6 = f'''
    SELECT
        experience,   -- тип требуемого опыта работы
        ROUND(AVG(    -- средняя зарплата, округлённая до целого чисоа
            CASE
                -- если заполнены оба поля с зарплатой
                WHEN salary_from IS NOT NULL AND salary_to IS NOT NULL
                THEN (salary_from + salary_to) / 2.0
                -- если заполнено только одно из полей
                ELSE COALESCE(salary_from, salary_to)
            END
        )) AS avg_salary
    FROM vacancies
    WHERE   -- Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой
    (
        salary_from IS NOT NULL
        OR salary_to IS NOT NULL
    )
    AND     -- условия отнесения вакансии к DS
    (
        LOWER(name) LIKE '%data scientist%'
        OR LOWER(name) LIKE '%data science%'
        OR LOWER(name) LIKE '%исследователь данных%'
        OR (
            LOWER(name) LIKE '%ml%'
            AND LOWER(name) NOT LIKE '%html%'  -- не нужно брать вакансии по HTML
        )
        OR LOWER(name) LIKE '%machine learning%'
        OR LOWER(name) LIKE '%машинн%обучен%'
    )
    GROUP BY experience  -- группируем по опыту работы
    ORDER BY
    CASE experience
        WHEN 'Нет опыта' THEN 1
        WHEN 'От 1 года до 3 лет' THEN 2
        WHEN 'От 3 до 6 лет' THEN 3
        WHEN 'Более 6 лет' THEN 4
        ELSE 5
    END;
'''
```

```python
# результат запроса
df = pd.read_sql_query(query_6_6, connection)
df
```

Дата-сайентист с опытом работы от 3 до 6 лет в среднем может рассчитывать на зарплату 256454.
