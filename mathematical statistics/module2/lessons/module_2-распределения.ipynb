```python
# PMF дискретного равномерного распределения
import numpy as np                          # импорт библиотеки numpy
from scipy.stats import randint             # импорт метода randint для работы с дискретным равномерным распределением
import matplotlib.pyplot as plt             # импорт функций matplotlib, реализующих графику
plt.figure(figsize=(15,10))                 # создание поля для графика с размерами 15х10
x = np.arange(0, 25, 0.001)                 # создание np-массива чисел от 0 до 25 с шагом 0.001
''' дальнейшие операции — изображение на поле точечного графика, где ось абсцисс — np-массив, 
ось ординат — PMF дискретного равномерного распределения в разных диапазонах, для каждого диапазона свои размер точек, 
цвет, маркер и название, в котором указано количество чисел в распределении'''
plt.scatter(x, randint.pmf(x, 1, 5), s = 60, c="blue", marker="o", label = 'n = 4')
plt.scatter(x, randint.pmf(x, 1, 8), s = 80, c="red", marker="*", label = 'n = 7')
plt.scatter(x, randint.pmf(x, 1, 16), s = 120, c="green", marker="+", label = 'n = 15')
plt.grid()                                  # создание сетки на графике
plt.legend(loc='upper right')               # размещение легенды в верхнем правом углу
plt.show()                                  # отображение графика
```

```python
# CDF дискретного равномерного распределения
import numpy as np                          # импорт библиотеки numpy
from scipy.stats import randint             # импорт метода randint для работы с дискретным равномерным распределением
import matplotlib.pyplot as plt             # импорт функций matplotlib, реализующих графику
plt.figure(figsize=(15,10))                 # создание поля для графика с размерами 15х10
x = np.arange(0, 25, 0.001)                 # создание np-массива чисел от 0 до 25 с шагом 0.001
''' дальнейшие операции — изображение на поле линейного графика, где ось абсцисс — np-массив, 
ось ординат — CDF дискретного равномерного распределения в разных диапазонах, для каждого диапазона свои
цвет и название, в котором указано количество чисел в распределении'''
plt.plot(x, randint.cdf(x, 1, 5), c="blue", label = 'n = 4')
plt.plot(x, randint.cdf(x, 1, 8), c="red", label = 'n = 7')
plt.plot(x, randint.cdf(x, 1, 16), c="green", label = 'n = 15')
plt.grid()                                  # создание сетки на графике
plt.legend(loc='lower right')               # размещение легенды в нижнем правом углу
plt.show()                                  # отображение графика
```

```python
# PMF распределения Бернулли
import numpy as np                          # импорт библиотеки numpy
from scipy.stats import bernoulli           # импорт метода bernoulli для работы с распределением Бернулли
import matplotlib.pyplot as plt             # импорт функций matplotlib, реализующих графику
plt.figure(figsize=(15,10))                 # создание поля для графика с размерами 15х10
x = np.arange(0, 1.001)                     # создание np-массива [0, 1]
''' дальнейшие операции — изображение на поле точечного графика, где ось абсцисс — np-массив, 
ось ординат — PMF распределения Бернулли, для каждого распределения свои размер точек, 
цвет, маркер и название, в котором указано значение p — вероятность успеха'''
plt.scatter(x, bernoulli.pmf(x, 0.3), s = 60, c="blue", marker="o", label = 'p = 0.3')
plt.scatter(x, bernoulli.pmf(x, 0.6), s = 80, c="red", marker="*", label = 'p = 0.6')
plt.scatter(x, bernoulli.pmf(x, 0.8), s = 120, c="green", marker="x", label = 'p = 0.8')
plt.grid()                                  # создание сетки на графике
plt.legend(loc='upper center')              # размещение легенды в центре вверху
plt.show()                                  # отображение графика
```

```python
# CDF распределения Бернулли
import numpy as np                          # импорт библиотеки numpy
from scipy.stats import bernoulli           # импорт метода bernoulli для работы с распределением Бернулли
import matplotlib.pyplot as plt             # импорт функций matplotlib, реализующих графику
plt.figure(figsize=(15,10))                 # создание поля для графика с размерами 15х10
x = np.arange(-0.5, 1.5, 0.001)             # создание np-массива 
''' дальнейшие операции — изображение на поле линейного графика, где ось абсцисс — np-массив, 
ось ординат — CDF распределения Бернулли, для каждого распределения свои цвет, маркер и название, 
в котором указано значение p — вероятность успеха'''
plt.plot(x, bernoulli.cdf(x, 0.3), c="blue", marker="o", label = 'p = 0.3')
plt.plot(x, bernoulli.cdf(x, 0.6), c="red", marker="*", label = 'p = 0.6')
plt.plot(x, bernoulli.cdf(x, 0.8), c="green", marker="x", label = 'p = 0.8')
plt.grid()                                  # создание сетки на графике
plt.legend(loc='lower center')              # размещение легенды в центре внизу
plt.show()
```
