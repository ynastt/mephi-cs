## 1
Импортёр упаковывает чай в пакеты с номинальным весом 125 грамм. Известно, что упаковывающая машина работает с известным среднеквадратическим отклонением равным 10 г. Случайным образом для проверки было выбрано 50 пакетов чая. Выборочное среднее этих пакетов составило 125,8 г. Постройте доверительный интервал уровня надёжности 0.95 для среднего веса пакета: $P\{?<a<?\}=0.95$

```python
n = 50
sigma = 10
alpha = 0.05
x0 = 125.8
z = abs(stats.norm.ppf(alpha/2))
print(f'({round(x0-sigma/n**0.5*z,2)} ; {round(x0+sigma/n**0.5*z,2)})')
```

```python
n1 = round((z*sigma/(2/2))**2)
print(n1)
round(x0+sigma/n1**0.5*z, 2) - round(x0-sigma/n1**0.5*z, 2)
```
