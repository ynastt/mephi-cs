По данным выборки объема n = 12 было найдено, что ∑▒x_i =216, ∑▒x_i^2 =4046. Постройте 90% доверительный интервал для теоретической дисперсии. 

```python
import numpy as np
import scipy.stats as stats
n = 12
sum_x = 216
sum_x_2 = 4046
alpha = 0.1
x_mean = sum_x/n
D = 1/(n-1)*(sum_x_2 - n*x_mean**2)
chi2_r = stats.chi2.ppf(1-alpha/2, df=n-1)
chi2_l = stats.chi2.ppf(alpha/2, df=n-1)
print(f'Доверительный интервал: ({round((n-1)*D/chi2_r,2)}; {round((n-1)*D/chi2_l,2)})')
```
