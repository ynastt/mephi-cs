```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
import seaborn as sns
import scipy.stats as stats
```

### Задание №1. Проведите корреляционный анализ данных из файла 'ex1.csv' по Пирсону

```python
# считывание файла
df1 = pd.read_csv('ex1.csv') 
df1
```

```python
# КРИТЕРИЙ 1

# проверка, что все элементы - численного типа
print(df1.dtypes)

# построение гистограмм распределения
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
histplot1 = sns.histplot(
    data=df1,
    x='x',
    bins=100,
    ax=axes[0]
)
histplot1.set_title('Распределение x', fontsize=16)
histplot1.set_xlabel('x')

histplot2 = sns.histplot(
    data=df1,
    x='y',
    bins=100,
    ax=axes[1]
)
histplot2.set_title('Распределение y', fontsize=16)
histplot2.set_xlabel('y')
```

```python
# построение "ящика с усами" для изучения выбросов
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))
box1 = sns.boxplot(x='x', data=df1, ax=axes[0])
box1.set_title('Ящик с усами для изучения выбросов x', fontsize=16)
box2 = sns.boxplot(x='y', data=df1, ax=axes[1])
box2.set_title('Ящик с усами для изучения выбросов y', fontsize=16)
plt.show()
```

```python
# построение диаграммы рассеяния
p = sns.scatterplot(data=df1, x='x', y='y')
p.set_title("Диаграмма рассеяния между x и y", fontsize=16)
```

```python
# вывод о целесообразности использования корреляции Пирсона с обоснованием на основе предшествующих графиков
print("Вывод: Целесообразно использовать корреляцию Пирсона, т.к. графики ранее показывают наличие линейной зависимости между численными переменными х и у, данные распределены нормально, наблюдений много")
```

```python
# КРИТЕРИЙ 2

# расчет коэффициента Пирсона
r, p_value = stats.pearsonr(df1['x'], df1['y'])
print(f"Коэффициент корреляции Пирсона: {r}")
print(f"p-значение: {p_value}")

# расчет t-критерия
n = df1.shape[0]
t_score = r*(n-2)**0.5 / (1-r**2)**0.5
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df = n-2)

print('Значение статистики t:', t_score)
print('Критическое значение статистики t:',t_critical)

## Принятие решения на основе критического значения t
if np.abs(t_score) > t_critical:
    print("Связь статистически значима")
else:
    print("Нельзя отклонить нулевую гипотезу")
    
## Принятие решения на основе p-значения
if p_value < alpha:
    print("Связь статистически значима")
else:
    print("Нельзя отклонить нулевую гипотезу")
```

```python
# вывод о степени корреляции величин и статистической значимости их взаимосвязи 
print("Коэффициент корреляции приблизительно равен 0.8, следовательно наблюдаем сильную прямую взаимосвязь и корреляцию между х и у. На основе t-критерия считаем, что связь статистически значимая.")
```

### Задание №3. Проведите линейный регрессионный анализ данных из файла 'ex3.csv'

```python
# считывание файла
df3 = pd.read_csv('ex3.csv') 
df3.head()
```

```python
# КРИТЕРИЙ 5

# построение диаграммы рассеяния
p = sns.scatterplot(data=df3, x='x', y='y')
p.set_title("Диаграмма рассеяния между x и y", fontsize=16)

# вывод о целесообразности использования корреляции Пирсона
# Целесообразно использовать корреляцию Пирсона, т.к. на диаграмме рассеяния видно наличие линейной обратной зависимости между численными переменными х и у.

# расчет коэффициента Пирсона
r, p_value = stats.pearsonr(df3['x'], df3['y'])
print(f"Коэффициент корреляции Пирсона: {r}")
print(f"p-значение: {p_value}")

# расчет t-критерия
n = df3.shape[0]
t_score = r*(n-2)**0.5 / (1-r**2)**0.5
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df = n-2)

print('Значение статистики t:', t_score)
print('Критическое значение статистики t:',t_critical)

## Принятие решения на основе критического значения t
if np.abs(t_score) > t_critical:
    print("Связь статистически значима")
else:
    print("Нельзя отклонить нулевую гипотезу")
    
## Принятие решения на основе p-значения
if p_value < alpha:
    print("Связь статистически значима")
else:
    print("Нельзя отклонить нулевую гипотезу")
```

Коэффициент пирсона приблизительно равен -1 (-0.99 при округлении до двух знаком псоле запятой), что говорит о сильной корреляции и об обратной линейной зависимости х и у, при этом согласно t-критерию связь является статисчтиески значимой

```python
# КРИТЕРИЙ 6

# расчет коэффициентов линейной регрессии
res = stats.linregress(df3['x'], df3['y'])
print(f'Коэффициент наклона {res.slope}, сдвиг {res.intercept}')
print(f'Коэффициент корреляции {res.rvalue}, коэффициент детерминации {res.rvalue**2}, p-value (из теста Вальда в предположении, что наклон равен нулю) {res.pvalue}')
print(f'Ошибка наклона {res.stderr}, ошибка сдвига {res.intercept_stderr}')
print(f' Y = {round(res.intercept, 3)} + {round(res.slope, 3)}*X')

# построение диаграммы рассеяния прогнозных и реальных значений на одном графике
y_pred = res.intercept + res.slope*df3['x']
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(df3['x'], df3['y'])
ax.scatter(df3['x'], y_pred)
plt.title("диаграмма рассеяния прогнозных и реальных значений")
plt.show()
```

```python
# КРИТЕРИЙ 7

# расчет средней ошибки аппроксимации
appr_err = np.abs((df3['y'] - (res.intercept + res.slope * df3['x']))/df3['y']).mean()
print(f'Средняя ошибка аппроксимации: {appr_err}')
```

Средняя ошибка аппроксимации не должна превышать 0.1(10%). В данном случае она получилась значительно больше, чем 0.1. 

```python
# КРИТЕРИЙ 8

# расчет F-критерия
from scipy.stats import f

F_score = (res.rvalue**2/(1 - res.rvalue**2))*(n-2)
F_critical = stats.f.ppf(1-alpha, 1, n - 1 - 1)
print(f'значение статистики F: {F_score}\nКритическое значение статистики F: {F_critical}')
pvalue_F = f.sf(F_score, 1, n - 2)
print(f'p-значение: {pvalue_F}')

# вывод о статистической значимости уравнения в целом
# принятие решения на основе критического значения F
if np.abs(F_score) > F_critical:
    print("Уравнение статистически значимо")
else:
    print('Уравнение статистически не значимо')
    
if pvalue_F < 0.05:
    print('Уравнение статистически значимо')
else:
    print('Уравнение статистически не значимо')
```

```python
# КРИТЕРИЙ 9

# расчет стандартных ошибок коэффициентов регрессии и коэффициента корреляции
print(f'Ошибка наклона {res.stderr}, ошибка сдвига {res.intercept_stderr}')

# расчет t-критерия коэффициентов регрессии и коэффициента корреляции
t_coef = abs(stats.t.ppf(alpha/2, n-2))
t_real = F_score**0.5
print(f'Расчетное значение t-критерия: {t_real}\nКритическое значение t-критерия: {t_coef}')

# вывод о статистической значимости коэффициентов регрессии
if np.abs(t_real) > t_coef:
    print("Коэффициенты регрессии статистически значимы")
else:
    print("Коэффициенты регрессии не являются статистически значимыми")
    
# запись диапазона коэффицентов регрессии
print(f"slope (95%): {res.slope:.3f} +/- {t_coef*res.stderr:.3f}")
print(f"intercept (95%): {res.intercept:.3f} +/- {t_coef*res.intercept_stderr:.3f}")
```

```python
# КРИТЕРИЙ 10

# Дать предсказание в виде интервальной оценки для y при x0 = 6
x0 = 6
y0 = res.intercept + res.slope * x0
var = (y_pred - df3['y'])**2
S0 = (var.sum()/(n-1-1))**0.5
delta = S0 * (1+1/n+(x0-df3['x'].mean())**2 / (n*df3['x'].var()))*t_coef
print(f'Значение в интервале ({y0-delta}; {y0+delta})')
```

### Задание №4. Проведите линейный регрессионный анализ данных из файла 'ex4.csv'

```python
# считывание файла
df4 = pd.read_csv('ex4.csv') 
df4.head()
```

```python
# КРИТЕРИЙ 11

# построение диаграмм рассеяния (x1/y, x2/y, x1/x2)
plt.figure(figsize=(6, 15))
plt.subplot(3, 1, 1)
plt.scatter(df4['x1'], df4['y'], color='blue')
plt.title('Диаграмма рассеяния: x1 и y')
plt.xlabel('x1')
plt.ylabel('y')

plt.subplot(3, 1, 2)
plt.scatter(df4['x2'], df4['y'], color='green')
plt.title('Диаграмма рассеяния: x2 и y')
plt.xlabel('x2')
plt.ylabel('y')

plt.subplot(3, 1, 3)
plt.scatter(df4['x1'], df4['x2'], color='red')
plt.title('Диаграмма рассеяния: x1 и x2')
plt.xlabel('x1')
plt.ylabel('x2')

plt.tight_layout()
plt.show()
```

```python
# вывод о корректности использования модели множественной регрессии
print("Вывод: Анализируя диаграммы рассеяния, можно предположить наличие обратной зависимости y x1/y,наличие прямой зависимости у x2/y и отсутствие корреляции у x1/x2. При таких условиях модель множественной регрессии будет уместна для использования.")

# расчет коэффициентов корреляции
print("\nМатрица корреляции")
corr_matrix = df4.corr()
print(corr_matrix)

# вывод о необходимости устранения какого-либо фактора
print("\nВывод: Нет необходимости исключать какой-либо фактор. По матрице корреляции видна умеренная корреляция переменных x1, x2 с переменной y. А корреляция между x1 и x2 низкая, т.е. мультиколлинеарности нет.")
```

```python
# КРИТЕРИЙ 12

# расчет стандартных коэффицентов регрессии
import statsmodels.api as sm

X = df4[['x1', 'x2']]
y = df4['y']
# стандартизируем данные
X_std = (X - X.mean()) / X.std()
y_std = (y - y.mean()) / y.std()

model = sm.OLS(y_std, X_std).fit()
print(model.summary())
coefficients = model.params
print(f"\nКоэффициенты стандартной регрессии: \n'{model.params.index[0]}' : {coefficients[0]}, \n'{model.params.index[1]}' : {coefficients[1]}")
```

```python
# КРИТЕРИЙ 13

# расчет коэффициентов "чистой" регрессии
X = df4[['x1', 'x2']]
y = df4['y']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(f"Коэффициенты чистой регрессии: \n'{model.params.index[0]}' : {model.params.iloc[0]}, \n'{model.params.index[1]}' : {model.params.iloc[1]}, \n'{model.params.index[2]}' : {model.params.iloc[2]}")

y_pred = model.predict(X)
# построение диаграмм рассеяния прогнозных и реальных значений на одном графике
plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
plt.title('Прогнозные vs Реальные значения')
plt.xlabel('Реальные значения (y)')
plt.ylabel('Прогнозные значения (y_pred)')
plt.grid()
plt.show()
```

```python
# КРИТЕРИЙ 14

# расчет индекса множественной корреляции
r_sqr = model.rsquared
corr_index = r_sqr ** 0.5
# print("Коэффициент детерминации (R^2):", r_sqr)
print("Индекс множественной корреляции (R):", corr_index)

# расчет F-критерия
F_score = model.fvalue
p_value_f = model.f_pvalue
print(f"F-критерий: {F_score}")
print(f"p-значение: {p_value_f}")

# вывод о статистической значимости уравнения в целом
if p_value_f < 0.05:
    print("Уравнение статистически значимо")
else:
    print("Уравнение статистически не значимо")
```

```python
# КРИТЕРИЙ 15

# построение 3D-изображения прогнозных и реальных значений на одном графике
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df4['x1'], df4['x2'], y, color='green', label='Реальные значения')
ax.scatter(df4['x1'], df4['x2'], y_pred, color='orange', label='Прогнозные значения')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('Распределение реальных и прогнозных значений на 3D-графике')
ax.legend()
plt.show()
```
