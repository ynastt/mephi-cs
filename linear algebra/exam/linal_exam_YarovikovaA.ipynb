### Задание 1
Исследуйте систему векторов на линейную зависимость или независимость без использования Python (вручную):  
$ \vec{a}_1 = (1; 8; -1), \ \vec{a}_2 = (-2; 3; 3), \ \vec{a}_3= (4; -11; 9) $  

Подробно распишите решение, а в итоге сделайте вывод про то, являются ли векторы зависимыми или независимыми.

Решение:  

Векторы $ \vec{a}_1, \vec{a}_2, \vec{a}_3 $ линейно зависимы, если существуют такие коэффициенты $c_1, c_2, c_3$, не все из которых одновременно равны нулю, что выполняется:

$
c_1 \vec{a}_1 + c_2 \vec{a}_2 + c_3 \vec{a}_3 = \vec{0}.
$

Если таких коэффициентов нет, то векторы линейно независимы.

***Составляем уравнения:***

Подставим векторы $ \vec{a}_1, \vec{a}_2, \vec{a}_3 $:

$
c_1 \cdot (1, 8, -1) + c_2 \cdot (-2, 3, 3) + c_3 \cdot (4, -11, 9) = (0, 0, 0).
$

Получим систему линейных уравнений:

$
\begin{aligned}
& \quad c_1 - 2c_2 + 4c_3 = 0, \\
& \quad 8c_1 + 3c_2 - 11c_3 = 0, \\
& \quad -c_1 + 3c_2 + 9c_3 = 0.
\end{aligned}
$

**Запишем в матричном представлении:**

$$
\begin{bmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{bmatrix}
\cdot
\begin{bmatrix}
c_1 \\
c_2 \\
c_3
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
$$

**Определение линейной зависимости/независимости:**

Чтобы определить, явлются ли вектора линейно зависимыми (т.е. существуют $(c_1, c_2, c_3) \neq (0, 0, 0)$), необходимо найти ранг составленной матрицы.

Найдем определитель матрицы:

$$
\text{detA}
=
\text{det}
\begin{bmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{bmatrix}
$$

Разложим по первой строке:

$
\text{detA} =
1 \cdot
\begin{vmatrix}
3 & -11 \\
3 & 9
\end{vmatrix}
-
(-2) \cdot
\begin{vmatrix}
8 & -11 \\
-1 & 9
\end{vmatrix}
+
4 \cdot
\begin{vmatrix}
8 & 3 \\
-1 & 3
\end{vmatrix}
$

$
\text{detA} = 1 \cdot (27 + 33) - (-2) \cdot (72 - 11) + 4 \cdot (24 + 3) = 1 \cdot 60 - (-2) \cdot 61 + 4 \cdot 27 = 60 + 122 + 108 = 290
$

$
\text{detA} \neq 0 ⇒ rang(A) = 3
$

Так как определитель матрицы не равен нулю, то ранг матрицы 3x3 равен 3 и является полным рангом, значит система уравнений иммет только тривиальное решение $(c_1, c_2, c_3) = (0, 0, 0)$. Значит данные векторы являются линейно независимыми.

**Ответ:**

Векторы $ \vec{a}_1, \vec{a}_2, \vec{a}_3 $ являются линейно независимыми.

$
c_1 \vec{a}_1 + c_2 \vec{a}_2 + c_3 \vec{a}_3 = \vec{0}.
$

Если таких коэффициентов нет, то векторы линейно независимы.

Подставим векторы $ \vec{a}_1, \vec{a}_2, \vec{a}_3 $:

$
c_1 \cdot (1, 0, 2, -3) + c_2 \cdot (2, 2, -4, a) + c_3 \cdot (3, 1, 2, -5) = (0, 0, 0, 0).
$

Запишем в матричном представлении:

$$
\begin{bmatrix}

\end{bmatrix}
\cdot
\begin{bmatrix}
c_1 \\
c_2 \\
c_3
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0 \\
0
\end{bmatrix}
$$

Линейная зависимость векторов будет означать, что один из векторов может быть представлен через линейную комбинацию других векторов. В таком случае ранг матрицы будет меньше числа векторов.

Чтобы проверить условие линейной зависимости, исследуем определитель подматрицы 3×3 из первых трех строк и столбцов:

$
\det \begin{bmatrix}

0 & 2 & 1 \\
2 & -4 & 2
\end{bmatrix} =  1\cdot 2 \cdot 2 + 0 \cdot 3 \cdot (-4) + 2 \cdot 1 \cdot 2 - 3 \cdot 2 \cdot 2 - 1 \cdot 1 \cdot (-4) - 2 \cdot 2 \cdot 0 = 4 + 0 + 4 - 12 + 4 - 0 = 4 \cdot 3 -12 = 12 - 12 = 0
$

Так как определитель подматрицы 3×3 равен нулю, то векторы $ \vec{a}_1, \vec{a}_2, \vec{a}_3 $  уже линейно зависимы вне зависимости от значения параметра $a$.

**Ответ:**  

система векторов $ \vec{a}_1, \vec{a}_2, \vec{a}_3 $ является линейно зависимой при любом значении параметра $a$

### Задание 6
Найдите сумму сингулярных чисел для матрицы:

$$A=\left(\begin{array}{ccc}-2 & 0 & 1 \\ 0 & -2 & 1 \\ 1 & 1 & -1\end{array}\right)$$

Найдите решение без использования Python (вручную).

Решение:

Найдем транспонированную матрицу $A^T$:
$$ A^T = \begin{pmatrix} -2 & 0 & 1 \\ 0 & -2 & 1 \\ 1 & 1 & -1 \end{pmatrix} $$

Посчитаем произведение $A^TA \ $ ($ AA^T = A^TA$):

$ A^TA = \begin{pmatrix} -2 & 0 & 1 \\ 0 & -2 & 1 \\ 1 & 1 & -1 \end{pmatrix} \begin{pmatrix} -2 & 0 & 1 \\ 0 & -2 & 1 \\ 1 & 1 & -1 \end{pmatrix} = \begin{pmatrix} 5 & 1 & -3 \\ 1 & 5 & -3 \\ -3 & -3 & 3 \end{pmatrix} $

Составим характеристическое уравнение и найдем собственные значения для $A^TA$:
$ \det(A^TA - λE) = \begin{vmatrix} 5-λ & 1 & -3 \\ 1 & 5-λ & -3 \\ -3 & -3 & 3-λ \end{vmatrix} = -λ^3 + 13λ^2 - 36λ = -λ(λ^2 - 13λ + 36) = -λ(λ - 9)(λ - 4)= 0 $


Поулчаем собственные значения:
$$ λ_1 = 0, \quad λ_2 = 9, \quad λ_3 = 4 $$

Вычислим сингулярные числа (это квадратные корни из собственных значений матрицы $A^TA$):
$$ σ_1 = \sqrt{0} = 0 $$ $$ σ_2 = \sqrt{9} = 3 $$ $$ σ_3 = \sqrt{4} = 2 $$

Найдем сумму сингулярных чисел:
$$ σ_1 + σ_2 + σ_3 = 0 + 3 + 2 = 5 $$

**Ответ:**

Сумма сингулярных чисел матрицы $A$ равна 5.

### Задание 8  

Исследуем данные на наличие линейной зависимости между столбцами. Найдите ранг матрицы данных. Сравните его с количеством столбцов, сделайте вывод. Если линейная зависимость есть - напишите, между какими столбцами и удалите один из них.

Распишите решение с помощью кода на языке Python и напишите текстом интерпретацию полученного вывода.


```python
import numpy as np
import pandas as pd

# Рассичтаем корреляцию между числовыми столбцами. В датасете числовыми являются все столбцы
correlation_matrix = data.corr()
print('Корреляционная матрица: ')
display(correlation_matrix)

# Выводим пары столбцов с сильной линейной зависимостью
print("\nСтолбцы с высокой корреляцией (>0.8):")
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i,j]) > 0.8:
            print(f"{correlation_matrix.columns[i]} - {correlation_matrix.columns[j]}: {correlation_matrix.iloc[i,j]:.3f}")

# Находим ранг матрицы
matrix_rank = np.linalg.matrix_rank(data.select_dtypes(include=[np.number]))
print("Ранг матрицы:", matrix_rank)

# cравним ранг с количеством столбцов
num_columns = data.select_dtypes(include=[np.number]).shape[1]
print("Количество столбцов:", num_columns)
```

На основе полученных данных из анализа корреляционной матрицы видно, что у некоторых столбцов значения корреляции между собой очень высоки (близки к единице). Более того, количество столбцов матрицы больше ранга матрицы, что указывает на линейную зависимость.

Между следующими столбцами обнаружена сильная зависимость:
- temp и atemp: корреляция 0.992 — зависимость между температурой и температурой по ощущениям.
- windspeed(mph) и windspeed(ms): корреляция 1.000 — полная зависимость, что объясняется тем, что скорость ветра измеряется в двух единицах.

Удалим столбцы:
- 'atemp', т.к. фактическая температура (temp) это более объективный признак
- 'windspeed(mph)', поскольку скорость ветра измеряется в метрах в секунду в международной системе единиц.

```python
# Удаляем зависимые столбцы
data_cleaned = data.drop(['atemp', 'windspeed(mph)'], axis=1)

print("Сравним ранг и количество столбцов после удаления сильно зависимых столбцов из матрицы")
print("Ранг матрицы:", np.linalg.matrix_rank(data_cleaned.select_dtypes(include=[np.number])))
print("Количество столбцов:", data_cleaned.select_dtypes(include=[np.number]).shape[1])
```

 После удаления этих столбцов, матрица данных имеет полный ранг. Теперь всё оптимально для продолжения анализа.

### Задание 10

Разделите данные на предикторы и целевую переменную. Обучите модель с помощью классической линейной регрессии. Линейную регрессию необходимо реализовать без готовых библиотек.

Оцените качество модели с помощью хотя бы двух метрик качества.

Распишите решение с помощью кода на языке Python и напишите текстом интерпретацию полученного вывода.

```python
# Разделяем данные на предикторы и целевую переменную
X = data_cleaned.drop(['cnt'], axis=1)
y = data_cleaned['cnt']

# Добавляем столбец для смещения (bias) - столец из единиц
X = np.column_stack([np.ones(len(X)), X])

# Строим и обучаем модель линейной регрессии по МНК с матричной формулой
# Находим коэффициенты линейной регрессии
W_hat = np.linalg.inv(X.T @ X) @ X.T @ y

# Делаем прогноз
predict = X @ W_hat

# Метрики качества
print("Метрики качества модели:")
mse = np.mean((y - predict) ** 2)    # Среднеквадратичная ошибка
print(f"MSE: {mse:.3f}")

ss_residual = np.sum((y - predict) ** 2)
ss_total = np.sum((y - np.mean(y)) ** 2)
r2 = 1 - (ss_residual / ss_total)   #r2_score
print(f"R2 score: {r2:.3f}")

```

MSE отражает среднеквадратичную ошибку предсказаний. Значение MSE указывает на присутствие ошибок в предсказаниях, что характерно для реальных данных.

Значение метрики R² = 0.797 показывает, что модель объясняет 79.7% изменчивости целевой переменной, т.е. существует линейная функциональная зависимость между признаками и целевой переменной. Значение метрики близко к 0.8, что говорит о хорошем качестве модели(идеально когда R² =1)
