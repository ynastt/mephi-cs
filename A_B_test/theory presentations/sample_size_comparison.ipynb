```python
data_pivot = pickle.load(open('C:/Users/diana.nazipova/Desktop/Курс AB - 23 июля - 2 августа/День 3/data_pivot.pkl', 'rb'))
data_pivot_ = pickle.load(open('C:/Users/diana.nazipova/Desktop/Курс AB - 23 июля - 2 августа/День 2/data_pivot_.pkl', 'rb'))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

sns.heatmap(data_pivot, cmap='Oranges', annot=True, fmt='g', ax=ax1)
sns.set(font_scale=1.5)
ax1.set_title('Расчет требуемого размера группы имитационно')
ax1.set_xlabel(r'$\mu_T - \mu_C$')
ax1.set_ylabel(r'$\sigma$')

sns.heatmap(data_pivot_, cmap='Oranges', annot=True, fmt='g', ax=ax2)
sns.set(font_scale=1.5)
ax2.set_title('Расчет требуемого размера группы формулой')
ax2.set_xlabel(r'$\mu_T - \mu_C$')
ax2.set_ylabel(r'$\sigma$')

plt.show()
```
