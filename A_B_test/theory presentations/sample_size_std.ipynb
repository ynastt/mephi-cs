# Демонстрация. Стандартное отклонение метрики

Демонстрация влияния величины стандартного отклонения на видимый размер эффекта. При прочих равных меньшее стандартное отклонение позволяет заметить меньший эффект.

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind, norm
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

```python
size = 100_000
lift = 20

plt.rcParams.update({'font.size': 18})
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

lim_left = -100
lim_right = 100

# high std
std_high = 40
a = np.random.normal(0, std_high, size)
b = np.random.normal(lift, std_high, size)
ax1.axvline(x=np.mean(a), color='Red')
ax1.axvline(x=np.mean(b), color='Green')
sns.histplot(a, bins=100,  kde=True, ax=ax1, linewidth=0, color='Red', label='A', alpha=0.4)
sns.histplot(b, bins=100,  kde=True, ax=ax1, linewidth=0, color='Green', label='B', alpha=0.4)
ax1.set_title(fr'$\sigma$ = {std_high}')
ax1.set_xlim([lim_left, lim_right])


# low std
std_low = 10
a = np.random.normal(0, std_low, size)
b = np.random.normal(lift, std_low, size)
ax2.axvline(x=np.mean(a), color='Red')
ax2.axvline(x=np.mean(b), color='Green')
sns.histplot(a, bins=100,  kde=True, ax=ax2, linewidth=0, color='Red', label='A', alpha=0.4)
sns.histplot(b, bins=100,  kde=True, ax=ax2, linewidth=0, color='Green', label='B', alpha=0.4)
ax2.set_title(fr'$\sigma$ = {std_low}, уменьшено в {std_high/std_low:.1f} раз(а)')
ax2.set_xlim([lim_left, lim_right])

ax1.legend()
ax2.legend()
plt.show()
```
